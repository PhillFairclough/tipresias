```python
import numpy as np

betting_df = pd.read_csv('../data/full_data.csv')
# Betting data only goes back to 2010, so dropping earlier years for consistency
betting_df = (betting_df[(betting_df['year'] >= 2010) & (betting_df['oppo_team'] != '0')]
                        .set_index(['team', 'year', 'round_number'], drop=False)
                        .sort_index()
                        [['point_spread', 'win_odds']])
match_dict = {
    'team': np.append(match_df['home_team'], match_df['away_team']),
    'year': np.append(match_df['year'], match_df['year']).astype('int'),
    'round_number': np.append(match_df['round_number'], match_df['round_number']),
    'home': np.append(np.ones(len(match_df)), np.zeros(len(match_df))),
    'points': np.append(match_df['home_points'], match_df['away_points']),
    'oppo_team': np.append(match_df['away_team'], match_df['home_points']),
    'oppo_points': np.append(match_df['away_points'], match_df['home_points']),
    'venue': np.append(match_df['venue'], match_df['venue']),
    'margin': np.append(match_df['home_margin'], match_df['home_margin'] * -1)
}
stacked_df = pd.DataFrame(match_dict).set_index(['team', 'year', 'round_number']).sort_index()

df = pd.concat([stacked_df, betting_df], sort=True)
df
```
